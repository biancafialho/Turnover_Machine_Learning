```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
turnover_df= pd.read_csv("turnover.csv")
turnover_df
```

```python
turnover_df.info()
```

```python
(turnover_df.isna().sum()/turnover_df.shape[0]).sort_values(ascending=False)
```

```python
turnover_df.shape
```

- As colunas 'QtdeHorasTrabalhadasPadrao' e 'Acima18Anos' não apresentam variação,sendo preenchidas com o valor 80 e a letra Y respectivamente;
- A coluna 'NumeroEmpregado' contém apenas o número de identificação do funcionário;

```python
turnover_df = turnover_df.drop(columns=['QtdeHorasTrabalhadasPadrao'])
turnover_df = turnover_df.drop(columns=['Acima18Anos'])
turnover_df = turnover_df.drop(columns=['NumeroEmpregado'])
```

```python
turnover_df.shape
```

- Atualizando o tipo de informação das colunas indicadas como "object";

```python
turnover_df.Turnover = turnover_df.Turnover.astype('category')        
turnover_df.ViagensTrabalho = turnover_df.ViagensTrabalho.astype('category') 
turnover_df.Departamento = turnover_df.Departamento.astype('category')    
turnover_df.Educacao = turnover_df.Educacao.astype('category')        
turnover_df.Genero = turnover_df.Genero.astype('category')          
turnover_df.Cargo = turnover_df.Cargo.astype('category')           
turnover_df.EstadoCivil = turnover_df.EstadoCivil.astype('category')     
turnover_df.HoraExtra = turnover_df.HoraExtra.astype('category')
```

```python
# Histograma
plt.figure(figsize=(12,6))
sns.countplot(x='Turnover', data=turnover_df, alpha=0.85)
plt.xlabel('Turnover', size=12)
plt.show()
```
